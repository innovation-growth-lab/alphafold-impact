```python
from transformers import pipeline

# Load your model and tokenizer into the pipeline
model_name = "OpenAlex/distilbert-base-cased-finetuned-topic-classification-title-abstract"
classifier = pipeline("text-classification", model=model_name, top_k=10)  # Adjust top_k as needed

# Example text to classify
text = "Mega-scale experimental analysis of protein folding stability in biology and design"

# Get predictions
predictions = classifier(text)

# Define your probability threshold
threshold = 0.2  # For example, 20%

# Filter predictions that have a score greater than the threshold
filtered_predictions = [p for p in predictions[0] if p['score'] > threshold]

# Print the filtered predictions
for prediction in filtered_predictions:
    print(f"Label: {prediction['label']}, Score: {prediction['score']:.4f}")
```
